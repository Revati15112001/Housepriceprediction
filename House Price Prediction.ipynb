# Read csv file

```python
df=pd.read_csv("C:\\Users\\revat\\Downloads\\House Price Prediction.csv")
df
```

#Export the data set

```python
df.drop('date', axis= 1, inplace=True)
```

```python
column_to_move = 'price'
new_position = 20
column=df[column_to_move]
df = df.drop(column_to_move, axis=1)
df = pd.concat([df.iloc[:, :new_position], column, df.iloc[:, new_position:]], axis=1)
df
```

```python
x=df.iloc[:, :-1].values
y=df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, random_state=101)
```

#Training and Testing data set

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestRegressor 
regressor= RandomForestRegressor (n_estimators = 100, random_state= 0) 
regressor.fit(x_train, y_train)
```

```python
y_pred= regressor.predict(x_test)
np.set_printoptions (precision=2) 
print (np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
Mean_absolute_error = mean_absolute_error (y_test,y_pred) 
print('Mean absolute_error:',Mean_absolute_error)

Mean_squared_error = mean_squared_error(y_test,y_pred)
print('Mean squared error:',Mean_squared_error)

R_squared = r2_score (y_test,y_pred) 
print('R_squared:',R_squared)
```
